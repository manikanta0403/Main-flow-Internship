```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Load the data
df = pd.read_csv('householdtask3.csv')

# Display the first few rows of the DataFrame to understand its structure
print(df.head())
```

```python
# Check the DataFrame structure
print(df.info())

# Display the first few rows to understand the content
print(df.head())
```

Create Bar and Line Charts

```python
# Bar chart for total households ('tot_hhs') by year
plt.figure(figsize=(10, 5))
plt.bar(df['year'], df['tot_hhs'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.title('Total Households by Year')
plt.show()
```

```python
# Line chart for income and expenditure by year
plt.figure(figsize=(10, 5))
plt.plot(df['year'], df['income'], marker='o', linestyle='-', color='blue', label='Income')
plt.plot(df['year'], df['expenditure'], marker='o', linestyle='-', color='red', label='Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Income and Expenditure by Year')
plt.legend()
plt.show()
```

Combine Bar and Line Charts

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

# Bar chart on the primary y-axis
bars = ax1.bar(df['year'], df['tot_hhs'], color='skyblue', label='Total Households')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Households', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Line chart on the secondary y-axis
ax2 = ax1.twinx()
line1, = ax2.plot(df['year'], df['income'], marker='o', linestyle='-', color='blue', label='Income')
line2, = ax2.plot(df['year'], df['expenditure'], marker='o', linestyle='-', color='red', label='Expenditure')
ax2.set_ylabel('Amount', color='grey')
ax2.tick_params(axis='y', labelcolor='grey')

# Legends
ax1.legend(handles=[bars, line1, line2], loc='upper left')

plt.title('Total Households, Income, and Expenditure by Year')
fig.tight_layout()
plt.show()
```
