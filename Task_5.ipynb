```python
df = pd.DataFrame(data)
```

```python
# Feature Engineering
df['age_group'] = pd.cut(df['age'], bins=[0, 40, 50, 60, 100], labels=['<40', '40-50', '50-60', '60+'])
df['chol_level'] = pd.cut(df['chol'], bins=[0, 200, 240, 1000], labels=['normal', 'borderline high', 'high'])
df['heart_rate_range'] = pd.cut(df['thalach'], bins=[0, 120, 160, 250], labels=['low', 'normal', 'high'])
```

```python
# Interaction Features
df['age_chol'] = df['age'] * df['chol']
df['exang_oldpeak'] = df['exang'] * df['oldpeak']
```

```python
# Feature Selection
X = df.drop(columns=['target'])
y = df['target']
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X.select_dtypes(include=[int, float]))
```

```python
# PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_scaled)
```

```python


# Read the uploaded CSV file into a DataFrame
filename = list(uploaded.keys())[0]
df = pd.read_csv('heart.csv')

# Assuming the last column is the target variable
X = df.iloc[:, :-1]  # Features (all columns except the last one)
y = df.iloc[:, -1]   # Target (last column)

# Feature Importance with Random Forest
rf = RandomForestClassifier()
rf.fit(X, y)
feature_importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)

```

```python
# Feature Selection with Lasso
lasso = LassoCV()
lasso.fit(X_scaled, y)

# Get the column names from X_scaled, NOT X
lasso_coef = pd.Series(lasso.coef_, index=[f"Feature {i}" for i in range(X_scaled.shape[1])]).sort_values(ascending=False)

```

```python
# Output results
print("PCA Components:")
print(pca.components_)
```

```python
print("\nRandom Forest Feature Importances:")
print(feature_importances)
```

```python
print("\nLasso Coefficients:")
print(lasso_coef)
```
